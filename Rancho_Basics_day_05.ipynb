# Operators

- Operators

Operators are special symbols that trigger some mathematical calculations.
- operators take some inputs (called operands)
- operators return some output (called result)
- based on the types of inputs & outputs, operators can be:
    - arithmetic operators
    - relational/comparison operators
    - logical operators

- Arithmetic Operators

    - Inputs are numbers
    - Output is number

    [`+`, `-`, `*`, `/`]

```python
a = 22
b = 7

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
```

- Relational Operators

    - Inputs are numbers
    - Output is boolean

    [`==`, `!=`, `<`, `>`]

```python
a = 22
b = 7

print(f"{a} == {b} = {a == b}")
print(f"{a} != {b} = {a != b}")
print(f"{a} <  {b} = {a < b}")
print(f"{a} >  {b} = {a > b}")
```

- Logical Operators

    - Inputs are booleans
    - Output is boolean

    [`not`, `or`, `and`]

```python
a = True
b = False

print(f"not {a} = {not a}")
print(f"not {b} = {not b}")
print()
print(f"{a} or  {b} = {a or b}")
print()
print(f"{a} and {b} = {a and b}")
```

- Type casting

It is the process of converting a value from one data type to another.

```python
a = input("Enter a number: ")
b = input("Enter a number: ")

print(type(a), type(b))
print(f"{a} + {b} = {a + b}")

a = int(a)
b = int(b)

print(type(a), type(b))
print(f"{a} + {b} = {a + b}")
```

- Conditions

Conditions are expressions that result in a boolean value (i.e. either True or False).
- These can be used to alter the flow of the program at runtime.

```python
a = int(input("Value of a = "))
b = int(input("Value of b = "))

a_is_positive = (a > 0)
b_is_positive = (b > 0)
both_positive = a_is_positive and b_is_positive

print(f"Is a positive?      {a_is_positive}")
print(f"Is b positive?      {b_is_positive}")
print(f"Are both positive?  {both_positive}")
```

# Flow of Control

- `if`, `elif`, `else` Statements

These statements allow us to take decisions at runtime, based on some conditions.

```python
condition = False

print(1)

if (condition):
    print(2)
    print(3)

print(4)
```

```python
condition = False

print(1)

if (condition):
    print(2)
else:
    print(3)

print(4)
```

```python
condition1 = False
condition2 = True
condition3 = True

print(1)

if (condition1):
    print(2)

elif(condition2):
    print(3)

elif(condition3):
    print(4)

else:
    print(5)

print(6)
```

- `while` Statement

This statement allows us to repeat certain code, based on some condition.

```python
condition = True

print(1)

while (condition):
    print(2)
    print(3)
    
print(4)
```

```python
count = 0

print(f"Before : {count}")

while (count < 5):
    print(f"Inside : {count}")
    count = count + 1

print(f"After : {count}")
```

- Binary Search

We have to search for a unknown number selected by a person in a given range (say 1 to 100).

After each guess, the other person will reply with either or these:
- my number is higher than your number
- my number is lower than your number
- my number is equal to your number

Based on the response, you have to make a new guess. The game continues until you find the correct number.

- **Approach 1**

    Check for each number in the given range, starting from the lowest to the highest, until you find the correct number.

    As you can see, this is going to be very slow.

```python
lower_bound = 1
upper_bound = 100

print(f"Think of a number between {lower_bound} and {upper_bound}")

guess = lower_bound
match_found = False
count = 0

while(not match_found):
    print(f"My guess is {guess}")
    count += 1
    response = input("Select response (< = >): ")
    if(response == "="):
        print("I won!")
        match_found = True
    guess = guess + 1

print(f"I took {count} guesses")
```

- **Approach 2**

    Instead of checking of all numbers from start till end, we check for the middle number. This will result in 3 cases:
    - it matches with the unknown number: Game over
    - it is lower the unknown number: in this case will eliminate the upper half
    - it is higher the unknown number: in this case will eliminate the lower half

    So just a simple guess will reduce the range by half.

```python
lower_bound = 1
upper_bound = 100

print(f"Think of a number between {lower_bound} and {upper_bound}")

match_found = False
count = 0

while(not match_found):
    guess = (lower_bound + upper_bound) // 2
    count += 1

    print(f"My guess is {guess}")
    response = input("Select response (< = >): ")
    if(response == "<"):
        upper_bound = guess - 1
    elif(response == ">"):
        lower_bound = guess + 1
    else:
        print("I won!")
        match_found = True

print(f"I took {count} guesses")
```
