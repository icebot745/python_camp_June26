# Basic IO Operations

- `print()` function

This function helps us to display something on the output screen.

```python
print("Hello")
```

When you try to print something:
- with quotes `eg: print("x")`, Python tries to print it as it is.
- without quotes `eg: print(x)`, Python tries to print it's value.
    - if the value is not defined, it gives an error.

```python
print("123")    # displays 123
print(123)      # displays 123
print()         # displays an empty line
print("abc")    # displays abc
print(abc)      # raises Error
```

- Variables

Variables in programming are used to store values. They are like containers that hold values.

Once created you can:
- read the current value in it
- write a new value in it

```python
xyz = 123       # creates a variable `xyz` with value 123

print("xyz")    # displays xyz
print(xyz)      # displays 123

xyz = 456       # changes the value of xyz to 456

print("xyz")    # displays xyz
print(xyz)      # displays 456
```

You can use these techniques to print multiple values within a single print statement.

```python
print("xyz =", xyz)     # multiple values separated by comma
print(f"xyz = {xyz}")   # f-strings
```

```python
name = "Rancho"
age  = 5

print("My name is", name, "I am", age, "years old")
print(f"My name is {name} I am {age} years old")
```

- `input()` function

This function helps us take input from the user while the program is running (at run-time) and save it in a variable.
- by default it saves the input as a string (more on that later).

```python
name = input("What's your name? ")
age = input("What's your age? ")

print(f"Hello {name}. You are {age} years old.")
```

- Data Types

Python support the following basic data types:
|Data type|Fullform|Description|
|---|---|---|
|`int`  | Integers               |For numbers without a fractional part|
|`float`| Floating point numbers |For numbers with a fractional part|
|`str`  | Strings                |For text based data|
|`bool` | Boolean                |For True/False values|
|`None` | None type              |For absence of any value|

- `type()` function

This function is used to find the data type of the value / variable.

```python
var = 12
print(type(12))
print(type(var))
```

```python
var = 3.0
print(type(3.4))
print(type(var))
```

```python
var = "Hello"
print(type("Hello"))
print(type(var))
```

```python
var = False
print(type(False))
print(type(var))
```

```python
var = None
print(type(None))
print(type(var))
```
